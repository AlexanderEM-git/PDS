#### 3. Filtro de media movil
1. Disene un filtro de media movil con 4 diferentes valores de L en el rango de 50 a 500, calcule la respuesta de la senal (obtenida en el punto 2.1.6) ante dicho filtro y grafique la senal resultante usando subplots (se puede utilizar el vector de tiempos generado anteriormente). ¿Con cual considera que se visualiza y se podria analizar mejor la informacion? Tip: Para calcular la salida se usa la convolucion:

Donde senal es el vector con la senal. ¿Por que se debe usar mode=’same’?
R: same, se toma la logitud mas grande de las dos y el resultado tendra ese tamaño tambien

```python
#filtro_m_movil:

def filtro_m_movil(L):
    coefs=np.ones(L)/float(L)
    num=np.zeros(L)
    num[0]=1.0/L
    num[-1]=-1.0/L
    den=np.zeros(L)
    den[0]=1
    den[1]=-1
    return coefs

l = [2,160,260,360]

senal = mi_año
senal_smooth=np.convolve(senal, filtro_m_movil(l[0]), mode='same')
senal_smooth1=np.convolve(senal, filtro_m_movil(l[1]), mode='same')
senal_smooth2=np.convolve(senal, filtro_m_movil(l[2]), mode='same')
senal_smooth3=np.convolve(senal, filtro_m_movil(l[3]), mode='same')

t_año = np.arange(0,len(mi_año),1)  #vector de retraso

plt.figure(figsize=(20,10)) #tamaño de la grafica
plt.subplot(2,2,1) #dos graficas en una fila
plt.plot(t_año,senal_smooth)
plt.xlabel('Tiempo',fontsize=18) #label x
plt.ylabel('Datos',fontsize= 18) #label y
plt.title("Respuesta señal al filtro con L = 60",FontSize=15) #titulo
plt.legend()
plt.grid()

plt.subplot(2,2,2) #dos graficas en una fila
plt.plot(t_año,senal_smooth1)
plt.xlabel('Tiempo',fontsize=18) #label x
plt.ylabel('Datos',fontsize= 18) #label y
plt.title("Respuesta señal al filtro con L = 160",FontSize=15) #titulo
plt.legend()
plt.grid()

plt.figure(figsize=(20,10)) #tamaño de la grafica
plt.subplot(2,2,3) #dos graficas en una fila
plt.plot(t_año,senal_smooth2)
plt.xlabel('Tiempo',fontsize=18) #label x
plt.ylabel('Datos',fontsize= 18) #label y
plt.title("Respuesta señal al filtro con L = 260",FontSize=15) #titulo
plt.legend()
plt.grid()

plt.subplot(2,2,4) #dos graficas en una fila
plt.plot(t_año,senal_smooth3)
plt.xlabel('Tiempo',fontsize=18) #label x
plt.ylabel('Datos',fontsize= 18) #label y
plt.title("Respuesta señal al filtro con L = 360",FontSize=15) #titulo
plt.legend()
plt.grid()

plt.show() #mostrar
```

¿Con cual considera que se visualiza y se podria analizar mejor la informacion?
R: podemos ver que con un L = 60, tiene una representacion mas aproximada a la señal original y a sus datos, y a su vez mas suave, entre mas alto peor el resultado, entre menor orden del filtro mejor el resultado
